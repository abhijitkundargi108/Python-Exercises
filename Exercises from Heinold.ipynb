##### 3.	WAP which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. Enter a number: 4

Enter a number: nine

Enter a number: 7

16	3	5.33333333333



```python
tot= 0
count=0
avg= 0
while True:
    num=input("Enter a number: ")
    try:
        if num[0].lower() =='d':
            break
        tot=tot+int(num)
        count= count+1
        avg=tot/count
    except:
        print("Invalid input")
    continue
print('Total is ', tot)
print('Count is ', count)
print('Average is ', avg)        
```

##### 9. Write a program that lets the user play Rock-Paper-Scissors against the computer. There should be five rounds, and after those five rounds, your program should print out who won and lost or that there is a tie.

```python
C_count = 0
H_count = 0

li = ['rock','paper','scissor']

r = 5
for i in range(r):
    H = input('rock or paper or scissor: ')
    C = random.choice(li)
    print(C)
    if H == C:
        H_count += 1
    else:
        C_count += 1
    r -= 1
print('Human:',H_count)
print('Computer:', C_count)

if H_count > C_count:
    print('You won')
elif H_count == C_count:
    print('Tie')
else:
    print('You lost')
```

##### 12. Write a program that asks the user to enter a value n, and then computes (1+ 1/2 + 1/3 + . . . + 1/n )- ln(n). The ln function is log in the math module.

```python
import numpy as np
n = int(input('Enter an integer: '))
base = 1

for i in range(0,n):
    if i < n:
        ss = base + 1/(i+1)
    l = ss - np.log(n)
print(l)
```

##### 13. Write a program to compute the sum 1 - 2 + 3 - 4 + . . . + 1999 - 2000.

```python
n = int(input('Enter an integer: '))
ds=0

for i in range(1,n):
    d = 1 / (i +1)
    ds+= d
    su = (1 + ds) - np.log(n)
    print(d)
print('total:',su)
```

##### 16. Write a program that computes the factorial of a number. The factorial, n!, of a number n is the product of all the integers between 1 and n, including n. For instance, 5! = 1 * 2 * 3 * 4 * 5 = 120.

```python
n = int(input('Enter number to calculate factorial: '))
fac = 1
for i in range(1,n+1):
    fac *= i
    
print('factorial of'+' '+str(n)+' '+ 'is: ',fac)
```

##### 21. Write a program that asks the user to enter a word and determines whether the word is a palindrome or not. A palindrome is a word that reads the same backwards as forwards.

```python
pal = input('Enter a word: ')

pal1= []
for i in pal:
    pal1.append(i)
    
pal = list(pal)
pal1.reverse()
if pal == pal1:
    print('This is a Palindrome')
else:
    print('This is not a Palindrome')
```

###### 22. At a certain school, student email addresses end with @student.college.edu, while professor email addresses end with @prof.college.edu. Write a program that first asks the user how many email addresses they will be entering, and then has the user enter those addresses. After all the email addresses are entered, the program should print out a message indicating either that all the addresses are student addresses or that there were some professor addresses entered.

```python
nu = int(input('how many addresses? : '))
p=0
s=0

emails =[]
for i in range(1, nu+1):
    a = input('Enter email address:')
    emails.append(a)
print(emails)

for em in emails:
    if 'prof' in em:
        p +=1
    elif 'student' in em:
        s += 1
print('Count of Professor addresses:'+' '+str(p))
print('Count of Student addresses:'+' '+str(s))
```
